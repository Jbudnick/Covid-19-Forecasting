```python
prev = covid_df.iloc[0]['cases']
new_cases = [prev]
for each in covid_df['cases'][1:]:
    new_cases.append(each - prev)
    prev = each
new_cases
covid_df['new_cases'] = new_cases
```

```python
covid_df
```

```python
plt.style.use('ggplot')
plt.plot(covid_df['new_cases'])
plt.xticks(rotation = 90)
plt.title('New Cases of COVID-19')
plt.savefig('Covid-19 New')
```
